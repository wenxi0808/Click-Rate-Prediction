```python
# Load your scaled data
df_train = pd.read_csv('/content/drive/MyDrive/AI ML Project/Train_data_Max.csv')
df_val = pd.read_csv('/content/drive/MyDrive/AI ML Project/Train_data_testing.csv')

# Prepare the datasets
X_train = df_train.drop('click', axis=1)
y_train = df_train['click']
X_val = df_val.drop('click', axis=1)
y_val = df_val['click']
```

```python
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss',colsample_bytree=0.7, learning_rate=0.1, max_depth=3, n_estimators=200, subsample=1.0)
```

```python
xgb.fit(X_train, y_train)
```

```python
pred = xgb.predict(X_val)
```

```python
classification_report(y_val, pred)
```

```python
confusion_matrix(y_val, pred)
```

```python
y_pred = xgb.predict(X_val)
y_pred_probs = xgb.predict_proba(X_val)[:, 1]

# Calculate metrics
log_loss_value = log_loss(y_val, y_pred_probs)
report = classification_report(y_val, y_pred, output_dict=True)
f1_score = report['weighted avg']['f1-score']
accuracy = accuracy_score(y_val, y_pred)
cm = confusion_matrix(y_val, y_pred)
fpr, tpr, thresholds = roc_curve(y_val, y_pred_probs)
roc_auc = auc(fpr, tpr)

# Print and plot metrics
print(f"Performance of XGb on Validation Dataset:")
print(f"Accuracy: {accuracy}\nLog Loss: {log_loss_value}\nF1 Score: {f1_score}")
print("Classification Report:")
print(pd.DataFrame(report).transpose())
print("Confusion Matrix:")
print(cm)
```
